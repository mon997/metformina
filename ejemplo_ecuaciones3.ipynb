## Ejemplos de como graficar en python 

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def f(s,t):
    a = 4
    b = 7
    n = s[0]
    c = s[1]
    dndt = a * n - (c/(c+1)) * b * n
    dcdt = (c/(c+1)) * n - c + 1
    return [dndt, dcdt]

t = np.linspace(0,20)
s0=[20,5]

s = odeint(f,s0,t)

plt.plot(t,s[:,0],'r--', linewidth=2.0)
plt.plot(t,s[:,1],'b-', linewidth=2.0)
plt.xlabel("t")
plt.ylabel("S[N,C]")
plt.legend(["N","C"])
plt.show()
```

```python
# function dy/dx = x + y/5.
func = lambda y,x : x + y/5.
# Initial condition
y0 = -3  # at x=0
# values at which to compute the solution (needs to start at x=0)
x = np.linspace(0, 4, 101)
# solution
y = odeint(func, y0, x)
# plot the solution, note that y is a column vector
plt.plot(x, y[:,0])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x=np.arange(-5,5)
y_1= -x*1.85 -x*1.88
y_2= x*1.85 -x*0.458 + x*4.13
```

```python
plt.figure()

plt.plot(x, y_1)  

plt.xlim(-5,5)
plt.ylim(-5,5)

plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')

plt.figure()

plt.plot(x, y_2)

plt.xlim(-5,5)
plt.ylim(-5,5)

plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')
```
